
# Program to convert dates in Pandas Dataframe to Datetime data

* Here I have a created a program that takes Event name, Date and Time as the input for 12 rows, then the date and the time of events have been converted to  pandas datetime. The day of the event has been added as a new column with the help of dt.day_name() and the number of days left for the event has been calculated by taking the difference of present date and the date of event. 

* If the date or time has not been given in a proper format, then by deafult it will be NaT as I have given errors ="coerce" which will by default fill in NaT values.

```python
import pandas as pd
```

```python
data = pd.DataFrame(columns = ["Event","Date","Time"])

for i in range(12):
    print("---------------------------------------------------")
    events = input("Enter the Event: ")
    date = input("Date of the Event: ")
    time = input("Timings of the Event: ")
    print("----------------------------------------------------")
    
    data1 = pd.DataFrame(data =[[events,date,time]], columns = ["Event","Date","Time"])
   
    data = pd.concat([data,data1], axis = 0)
data["Date"]= pd.to_datetime(data["Date"],errors = 'coerce')
data["Time"]= pd.to_datetime(data["Time"],errors = 'coerce') .dt.time  
data["Day"] = data['Date'].dt.day_name()

data["DaysLeft"] =(data["Date"] - pd.to_datetime("today")).dt.days




data.index = range(len(data.index))
print("\n***24th Intercollegiate Fest 2022***")
display(data)   
```
