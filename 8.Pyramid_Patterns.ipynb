
# Python Program to create Pyramid Patterns

```python
def user_pattern():
    nrow = int(input("Enter the number of rows: "))
    pat = input("Pattern: ")
    for i in range(1,nrow+1):
        for c in range(i,nrow):
            print(end = " ")
        for j in range(1,i+1):
            print(pat, end =" ")
        print("\r")

def ascii_pattern():
    ascii_number = 65
    nrow = int(input("Enter the number of rows: "))
    for i in range(1,nrow+1):
        for j in range(i, nrow):
            print(end = " ")
        for k in range(1,i+1):
            character = chr(ascii_number)
            print(character, end=' ')
            ascii_number += 1
        print("\r")

def num_pattern():
    nrow = int(input("Enter the number of rows: "))

    for i in range(1,nrow+1):
        for c in range(i,nrow):
            print(end = " ")
        for j in range(1,i+1):
            print(i-1, end =" ")
        print("\r")

    nrow = int(input("Enter the number of rows for second pattern: "))
    num = 1
    for i in range(1,nrow+1):
        for c in range(i,nrow):
            print(end = " ")
        for j in range(1,i+1):
            print(num, end =" ")
            num = num+1
        print("\r")

while True:
    print("\n*****PYRAMID PATTERNS******\n")
    print("1.Pyramid with different Pattern\n2.Pyramid with Ascii Characters\n3.Pyramid with Numbers.\n************************************\n")
   
    try:
        choice = int(input("Enter your choice:"))
    
        if choice == 1:
            user_pattern()
        elif choice ==2:
            ascii_pattern()
        elif choice ==3:
            num_pattern()
        else:
            print("Invalid Choice!! Thank You")
            break
            
    except ValueError:
        print("Only Integer Values are allowed!!")

       
```
