# Program to iterate over rows and columns in a Pandas Dataframe


```python
import pandas as pd
```

```python
data = pd.read_csv("orders.csv", header = 0)
display(data)
data.describe()
data.head()
```

## Iterating over rows using index method

## To Iterate over the Date and Destination

```python
for i in data.index:
    print(data["Date"][i] , data["Destination"][i])
```

## Using loc[ ] function

## To Iterate over Driver name and the destination

```python
for i in range(len(data)):
    print(data.loc[i, "Driver_name"],"-", data.loc[i, "Destination"])
data.groupby(["Driver_name", "Destination"]).Destination.count()
```

## To obtain the Items which have  quantity more than 25 units

```python
data.loc[data["Number_of_items"]>25, ["Item"]] 
```

## To select rows which have airplane as the mode of transport

```python
data.loc[lambda data: data["Transport"]=="airplane"]
```

## Using iloc [] 

## To iterate over first ten rows

```python
data.iloc[:10]
```

## To obtain the first and second row corresponding to the Driver Name and Date

```python
data.iloc[[1,2], [2,1]]
```

## To obtain the value at the zeroth row and 1 column

```python
data.iloc[0,1]
```

## Using iterrows()

```python
print("---------To iterate over Date and Destination-----------\n")

for index, row in data.iterrows():
    print(row["Date"], row["Destination"])
print()


print("---------To iterate over Items and Transport---------\n")
for index, row in data.iterrows():
    print((row["Item"],":",row["Transport"]))
data.groupby(["Transport"]).Transport.count()

print()
print("---------To iterate over Driver and Destination---------\n")
for row in data.itertuples():
    print(getattr(row, "Driver_name"), ":",getattr(row, "Destination"))
print()

print("---------To iterate over Items>=20 and Item =TV---------\n")
for row in data.itertuples():
    d1 = row.Number_of_items>=20
    d2 = row.Item=="TV"
    print(row.Number_of_items,":",row.Item,":",d1&d2)
print()  
```

```python
for i ,j in data.iteritems():
    print("Column: ",i,)
    print("Value", j.values[1:5])
```
