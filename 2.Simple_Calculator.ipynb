
# Program to Generate a Simple Calculator

```python
def add(x,y):
    return x+y

def sub(x,y):
    return x-y

def mul(x,y):
    return x*y

def div(x,y):
    return x/y

def sroot(x,y):
    return x**0.5, y**0.5

def power(x,y):
    return x**y


import math
def logth(x,y):
    print("Input 1: Number, and Input 2: Base")
    return math.log(x,y)


while True:
      
    print("\n******SIMPLE CALCULATOR******\n")
    print("\n1.Addition\n2.Subtraction\n3.Multiplication\n4.Division\n5.Square Root\n6.Powern9.\n7.Log Value")
    choice =  input("\nSelect an Operation: ")
    if choice in ("1","2","3","4","5","6","7"):
        num1 = float(input("Input 1: "))
        num2 = float(input("Input 2: "))

        if choice == '1':
            print("You Chose Addition")
            print("Output:",num1 , "+", num2, "=", '%.2f'%add(num1,num2))

        elif choice =="2":
             print("You Chose Subtraction")
             print("Output:", num1 , "-", num2, "=", '%.2f'%sub(num1,num2))

        elif choice =="3":
            print("You Chose Multiplication")
            print("Output:",num1 , "*", num2, "=", '%.2f'%mul(num1,num2))

        elif choice =="4":
            print("You Chose Division")
            if num2 == 0:
              print("Can't divide by a number by 0")
              print("Please enter a valid number: ")
              num_2 = float(input("Input 2: "))
              print( "Output:",num1 , "/", num_2, "=", '%.2f'%div(num1,num_2))
            else:
              print( "Output:",num1 , "/", num2, "=", '%.2f'%div(num1,num2))

    

        elif choice =="5":
            print("You Chose Square Root")
            print("The square root of ", num1 , "and ", num2, "is: ", sroot(num1, num2))

        elif choice =="6":
            print("You Chose Power")
            print("Output:" ,num1, "^", num2, "= ", '%.2f'%power(num1,num2))


        elif choice == "7":
            print("You Chose Logarithmic Value")
            print("Output:", "Log({},{})".format(num1,num2), '%.2f'%logth(num1,num2))


    else:
        print("Invalid Operation Chosen!!!\Thank you")
        break


```
