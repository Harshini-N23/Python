# Python program to select or filter rows in a dataframe based on column values in Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data =pd.read_csv("sales.csv", index_col=0)
```

```python
del data["City"]
```

```python
del data["Tax 5%"]
```

```python
del data["cogs"]
```

```python
#del data["gross_margin percentage"]
del data["Time"]
```

```python
for i in data.index:
     branch = data["Branch"].value_counts()
     customer =    data["Customer_type"].value_counts()
     Gender = data["Gender"].value_counts()
     Product = data["Product_line"].value_counts()
     Payment = data["Payment"].value_counts()
print("Branch\n",branch)
print("\nCustomer type:\n", customer)
print("\nGender\n", Gender)
print("\nProduct Type:\n", Product)
print("\nPayment Mode\n", Payment)
```

## Branches having Gross Income greater above the average gross income

```python
avg_inc= data[data["gross_income"]>15]
display(avg_inc)

avg_inc.groupby(['Branch']).mean().plot(kind='pie', y = "gross_income", autopct ="%1.0f%%")
```

## Customers who have total price >= 322 

```python
totalprice = data[(data["Total"]>= np.mean(data["Total"])) & (data["Total"]<=max(data["Total"]))]
display(totalprice)

```

```python
print(totalprice.groupby(["Gender", "Payment"]).Gender.count())
plt.figure(figsize=(12,4))
sns.countplot(x = "Gender",  data = totalprice, hue ="Payment", palette ="Set1")
plt.show()
```

##  Electronic Accessories Purchased by Male and Female

```python
Male_EA = data[ (data["Product_line"]=="Food and beverages") & (data["Gender"] =="Male")]
display(Male_EA)
Fem_EA = data[(data["Product_line"]=="Food and beverages") & (data["Gender"] =="Female")]
display(Fem_EA)

EA = [Male_EA["Gender"].count(), Fem_EA["Gender"].count()]
plt.pie(EA,  autopct ="%1.0f%%" , labels = ["Male", "Female"])
plt.title("Percentage of Food and Beverages consumed by Male and Female")
plt.legend()
plt.show()
```

## Sales during the month of January

```python
date = data[(data["Date"]>="1/1/2019") & (data["Date"]<="1/31/2019")]

display(date.sort_values(by = "Date"))
print("The average income during the month of January: ", date["gross_income"].mean())
print("The product that was sold the highest: ", date["Product_line"].mode()[0])
```
