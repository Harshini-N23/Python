# Python Program to append rows in pandas dataframes using for loop

```python
sci = pd.read_csv("science.txt",header = 0)
display(sci)
com = pd.read_csv("comm.txt", header =0)
display(com)

l = list(sci)
result =[]
x = 0
for i in range(3):
    l2 = list(com.iloc[x])
    z =zip(l,l2)
    zd  = dict(z)
    result.append(zd)
    x = x+1

admin = sci.append(result).set_index(["Admission Number"])
display(admin)

```

```python
for i in range(0, len(admin.index),1):
    if admin["10th Marks"][i]>100.00 or admin["10th Marks"][i]<0:
        admin["10th Marks"][i] = None
    if admin["Year of Passing"][i]<2021 or admin["Year of Passing"][i]>2021:
        admin["Year of Passing"][i] = 0    
```

```python
admin["Result"] = "Null"
for i in range(0, len(admin.index),1):
    if admin["10th Marks"][i]> 80 :
        admin["Result"][i] = "Distinction"
    elif admin["10th Marks"][i]> 65:
        admin["Result"][i] = "First Class"
    elif admin["10th Marks"][i]> 45:
        admin["Result"][i] = "Second Class"
    elif admin["10th Marks"][i]> 35:
        admin["Result"][i] = "Pass"
    else:
        admin["Result"][i] = "--"
```

```python
admin  =admin.sort_values(by = 'Admission Number')
```

```python
admin.to_csv("university.txt")
```
