
## Score Grade Calculator

```python
from tabulate import tabulate

print("******SCORE GRADE CALCULATOR FOR MARKS OUT OF 100******\n")
print("Grading System:\n")

table = [['Percentage of Marks', "Eligible Grade"], ["From 90 to 100", "A+"], ["From 80 up-to 90", "A"], ["From 70 up-to 80", "B+"], ["From 60 up-to 70", "B"], ["From 50 up-to 60","C+"], ["From 35 up-to 50", "C"]]
print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))

s_name= input("Enter Your Name: ")
Class = (input("Class: " ))
Reg = int(input("Register Number: "))


print("\nEnter the Marks of Subjects\n")

    
marks=[]
tot = 0
while int(len(marks)+1)!=7:
    print("Subject",int(len(marks)+1), "= ", end= "")
    name = float(input())
    if name <0 or name>100:
        print("Enter valid marks")
        continue
    marks = marks+ [name]
    
    

print()
print("---------------------------------")
print("Marks"+" "+"Grade")
for name in marks:
    if name>=91 and name<=100:
        print(name, "A+")
    elif name>=81 and name<=90:
        print(name, "A")
    elif name>=71 and name <=80:
        print(name, "B+")
    elif name>=61 and name<=70:
        print(name, "B")
    elif name>=51 and name<=60:
        print(name, "C+")
    elif name>=41 and name<=50:
        print(name, "C")
    else:
        print(name, "Fail")
    
    tot = tot + name
print("---------------------------------")
print("\n********************************")
print("FINAL REPORT")
per = (tot/600)*100
table = [['Name', 'Class', 'Register', "Total Marks", "Total_Marks Obtained", "Percentage"], [s_name, Class, Reg,"600",tot, "{0:.2f}%".format(per)]]
print(tabulate(table))

if per>90 and per<=100:
    print("Grade : A+","\nRemarks: Outstanding")
elif per>80 and per<=90:
    print("Grade: A","\nRemarks: Excellent")
elif per>70 and per<=80:
    print("Grade:  B+","\nRemarks: Very Good")
elif per>60 and per<=70:
    print("Grade:  B","\nRemarks: Good")
elif per>50 and per<=60:
    print("Grade: C+","\nRemarks: Above Average")
elif per>35 and per<=50:
    print("Grade: C","\nRemarks: Average")
else:
    print("Fail","\nRemarks: Needs Improvement")
print("********************************")


    
```
